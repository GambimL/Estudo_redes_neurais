### Importando dados para estudo

Os dados utlizados são providos pelo iris data sete que possui um conjunto de 50 amostras de tres espécies 

```python
from sklearn import datasets

iris = datasets.load_iris(return_X_y = True)
x = iris[0]
y = iris[1]
y

datasets.load_iris()['feature_names']
datasets.load_iris()['target_names']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 

sns.scatterplot(x=x[:, 2], y=x[:, 3], hue = y, palette='tab10')
plt.xlabel('comprimento (cm)',fontsize =16)
plt.ylabel('largura (cm)', fontsize=16)
plt.title('Distribuição pétalas', fontsize = 18)
plt.show()
```

```python
sns.scatterplot(x = x[:,0], y = x[:,1], hue = y, palette = "tab10")
plt.xlabel('comprimeto (cm)', fontsize = 16)
plt.ylabel('largura (cm)', fontsize = 16)
plt.title('Distribuição sépalas', fontsize = 18)
plt.show()
```

```python
y = keras.utils.to_categorical(y)
y.shape
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
print(x)
```

```python
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 42)
y_teste
```

```python
modelo = keras.Sequential([
    keras.layers.InputLayer(input_shape=[4,], name = 'Entrada'),
    keras.layers.Dense(512, activation = 'relu', name='oculta',
                      kernel_initializer = keras.initializers.RandomNormal(seed = 142)),
    keras.layers.Dense(3, activation='softmax', name='saida')
    
])

modelo.summary()
```

```python

modelo.compile(loss = 'categorical_crossentropy',
               optimizer = 'rmsprop',
               metrics = ['categorical_accuracy'])
```

```python
epocas=100
historico = modelo.fit(x_treino, y_treino, epochs=epocas,  validation_split=0.3)
```

```python
perda, acuracia = modelo.evaluate(x_teste, y_teste)
print(f"perda: {perda}")
print(f"acuracia: {acuracia}")
```

```python
import numpy as np
teste = modelo.predict(x_teste)
print('resultado teste:', teste[1])
print('resultado y_teste:', y_teste[1])
```

```python
import pandas as pd 
plt.figure(figsize = (14, 5))
plt.subplot(1, 2, 1)
plt.plot(historico.history['loss'], color='red')
plt.plot(historico.history['val_loss'], color='green')
plt.grid()
plt.legend(['treino', 'validação'])
plt.xlabel('epocas')
plt.ylabel('perda')
plt.subplot(1, 2, 2)
plt.plot(historico.history['categorical_accuracy'], color='red')
plt.plot(historico.history['val_categorical_accuracy'], color='green')
plt.grid()
plt.legend(['treino', 'validação'])
plt.xlabel('epocas')
plt.ylabel('acuracia')
```
